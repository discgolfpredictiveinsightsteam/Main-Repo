## Examine Weather Data at KCASANFR1388

### Objective

*  Determination of whether or not station KCASANFR1388 provides similar data to other stations.

### Rationale

*  Why This?  All other stations were verified by weather measurements on course.  Since we know that four other stations are representative, but could not determine whether or not KCASANFR1388 was representative due to an outage on the test date, we can infer the usefulness of KCASANFR1388 by seeing wehter or not its data is similar to that of the other stations.

*  Why Me?  As the chief architect of the weather data system, I am the logical choice.

*  Why Now?  This is the final piece of information required to fully investigate all weather stations of relevance, and is a high priority unmet requirement prior to deployment

### Import / Set-Up

```python
import pandas as pd
import glob
import re
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get all weather station filenames for Golden Gate Park (KCASANFR)
filenames = glob.glob('../../wx_record/wx_station_by_date/KCASANFR*.csv')
```

```python
# Import data into DataFrame
wx_df = pd.DataFrame()
for filename in filenames:
    temp_wx_df = pd.read_csv(filename, parse_dates = ['time'])
    station_name = 'KCASANFR' + re.split('KCASANFR',filename)[1].split('_')[0]
    temp_wx_df['station'] = station_name
    wx_df = wx_df.append(temp_wx_df)
wx_df.head()
```

```python
# Clean up (set index to time)
wx_df = wx_df.drop(columns = ['Unnamed: 0'])
wx_df = wx_df.set_index('time')
wx_df.head()
```

```python
# Split into two DataFrames, one for KCASANFR1388 and one for the rest
k1388_df = wx_df[wx_df['station'] == 'KCASANFR1388']
rest_df = wx_df[wx_df['station'] != 'KCASANFR1388']
k1388_df.head()
```

```python
# Resample on one hour time-frame
rk1388_df = k1388_df.resample('60T').agg({'T':'mean','w_dir':'mean','w_gust':'mean','precip':'mean'})
rrest_df = rest_df.resample('60T').agg({'T':'mean','w_dir':'mean','w_gust':'mean','precip':'mean'})
rdrest_df = rest_df.resample('60T').agg({'T':'std','w_dir':'std','w_gust':'std','precip':'std'})
rk1388_df.head()
```

```python
merged_df = rk1388_df.merge(rrest_df,left_index=True, right_index=True, how='inner')
merged_df = merged_df.merge(rdrest_df,left_index=True, right_index=True, how='inner')
merged_df.head()
len(merged_df)
```

```python
# Now do a statitical test via z-score 
merged_df['Tz'] = (merged_df['T_x'] - merged_df['T_y']) / merged_df['T']
merged_df['wdz'] = (merged_df['w_dir_x'] - merged_df['w_dir_y']) / merged_df['w_dir']
merged_df['wgz'] = (merged_df['w_gust_x'] - merged_df['w_gust_y']) / merged_df['w_gust']
merged_df['pz'] = (merged_df['precip_x'] - merged_df['precip_y']) / merged_df['precip']
merged_df = merged_df.replace(np.inf, np.nan)
merged_df = merged_df.dropna()
len(merged_df)
```

```python
plt.hist(merged_df['Tz'], bins=30)
```

```python
plt.hist(merged_df['wdz'], bins=30)
```

```python
plt.hist(merged_df['wgz'], bins=90)
```

```python
plt.hist(merged_df['pz'], bins=30)
```

### Conclusions:

*  Temperature is consistently higher at KCASANFR1388, but only by 1 deg F -- not enough to matter
*  Winds are consistent in direction and gust
*  Precip is a bit lower, but within the range of other station variations
